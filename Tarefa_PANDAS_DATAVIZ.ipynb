```python
import pandas
```

```python
titanic = pandas.read_csv("./titanic.csv")
```

```python
titanic.head()
```

```python
#Número de passageiros
num_passageiros = titanic.shape[0]
print(f"O número total de passageiros é {num_passageiros}")
```

```python
#Número de mulheres a bordo
num_mulheres = titanic[titanic["Sex"] == "female"].shape[0]
print(f"O número total de mulheres a bordo é {num_mulheres}")
```

```python
#Número de homens a bordo
num_homens = titanic[titanic["Sex"] == "male"].shape[0]
print(f"O número total de homens a bordo é {num_homens}")
```

```python
#Média de idade dos passageiros
media_idade_passageiros = titanic["Age"].mean()
print(f"A média de idade dos passageiros é {media_idade_passageiros}")
```

```python
#Média de idade das mulheres a bordo
mulheres = titanic["Sex"] == "female"
media_idade_mulheres = titanic[mulheres]["Age"].mean()
print(f"A média de idade das mulheres a bordo é {media_idade_mulheres}")
```

```python
#Média de idade dos homens a bordo
homens = titanic["Sex"] == "male"
media_idade_homens = titanic[homens]["Age"].mean()
print(f"A média de idade dos homens a bordo é {media_idade_homens}")
```

```python
#Número de homens a bordo com idade acima de 45 anos
idade_acima_45 = titanic["Age"] > 45
numero_homens_acima_45 = titanic[homens & idade_acima_45].shape[0]
print(f"O número de homens a bordo com idade acima de 45 anos é {numero_homens_acima_45}")
```

```python
#Número de mulheres a bordo com idade abaixo de 25 anos
idade_abaixo_25 = titanic["Age"] < 25
numero_mulheres_abaixo_25 = titanic[mulheres & idade_abaixo_25].shape[0]
print(f"O número de mulheres a bordo com idade abaixo de 25 anos é {numero_mulheres_abaixo_25}")
```

```python
#Número de homens na primeira, segunda e terceira classe
homens_1_2_e_3_classe = titanic[homens]["Pclass"].value_counts().sort_index()
print(f"Número de homens na primeira, segunda e terceira classe:\n{homens_1_2_e_3_classe}")
```

```python
#Número de mulheres na primeira, segunda e terceira classe
mulheres_1_2_e_3_classe = titanic[mulheres]["Pclass"].value_counts().sort_index()
print(f"Número de mulheres na primeira, segunda e terceira classe:\n{mulheres_1_2_e_3_classe}")
```

```python
import matplotlib.pyplot as plt
```

```python
#Crie um histograma demonstrando as idades da primeira, segunda e terceira classe (informar a legenda de cores)
primeira_classe = titanic[titanic["Pclass"] == 1]["Age"]
segunda_classe = titanic[titanic["Pclass"] == 2]["Age"]
terceira_classe = titanic[titanic["Pclass"] == 3]["Age"]

plt.hist([primeira_classe, segunda_classe, terceira_classe], bins=20, stacked=True)
plt.legend(["Primeira Classe", "Segunda Classe", "Terceira Classe"])
plt.xlabel('Idade')
plt.ylabel('Número de Passageiros')
plt.title('Distribuição de Idades por Classe')
plt.show()
```

```python
homens_primeira_classe = titanic[(titanic["Sex"] == "male") & (titanic["Pclass"] == 1)]["Age"]
mulheres_primeira_classe = titanic[(titanic["Sex"] == "female") & (titanic["Pclass"] == 1)]["Age"]

plt.hist([homens_primeira_classe, mulheres_primeira_classe], bins=20, alpha=0.5)
plt.legend(["Homens", "Mulheres"])
plt.xlabel('Idade')
plt.ylabel('Número de Passageiros')
plt.title('Distribuição de Idades por Gênero na Primeira Classe')
plt.show()
```

```python
homens_segunda_classe = titanic[(titanic["Sex"] == "male") & (titanic["Pclass"] == 2)]["Age"]
mulheres_segunda_classe = titanic[(titanic["Sex"] == "female") & (titanic["Pclass"] == 2)]["Age"]

plt.hist([homens_segunda_classe, mulheres_segunda_classe], bins=20, alpha=0.5)
plt.legend(["Homens", "Mulheres"])
plt.xlabel('Idade')
plt.ylabel('Número de Passageiros')
plt.title('Distribuição de Idades por Gênero na Segunda Classe')
plt.show()
```

```python
homens_terceira_classe = titanic[(titanic["Sex"] == "male") & (titanic["Pclass"] == 3)]["Age"]
mulheres_terceira_classe = titanic[(titanic["Sex"] == "female") & (titanic["Pclass"] == 3)]["Age"]

plt.hist([homens_terceira_classe, mulheres_terceira_classe], bins=20, alpha=0.5)
plt.legend(["Homens", "Mulheres"])
plt.xlabel('Idade')
plt.ylabel('Número de Passageiros')
plt.title('Distribuição de Idades por Gênero na Terceira Classe')
plt.show()
```
